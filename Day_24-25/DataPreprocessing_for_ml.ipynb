```python
# importing libs for ml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("D:\Python ka Chilla Dobara\Day24-25\drawndata1.csv")
# df.head()

# df.sample(5) selects 5 random samples
df.sample(5)
```

## _Putting first two columns in Data as independent variables Just like in house example the size of the house was independent variable_
## _Putting last column as dependent variables just like the price of the house was dependent variable_

```python
# simply I am putting first two columns in X_in, and last columns in y_in
X_in = df[['x','y']]
y_in = df['z']          # this y_in contains both a and b
```

## _But in the code below I am only putting 'a' in "y_with_only_a" to plot in different color_

## _then in second line I am only putting 'b' in "y_with_only_b" to plot in different color_

```python
y_with_only_a = df['z']=='a'        # this contains only 'a'
y_with_only_b = df['z']=='b'        # this contains only 'b'
```

## _The below plot is scattered on raw data with 'a' in color_

```python
# on pure raw data
plt.scatter(X_in['x'],X_in['y'], c=y_with_only_a)

# the c is used for color, c=y_with_only_a it
# means that color the data that is inside of variable
# and we know that this variable contains only 'a' point
# you can see results in graph that 'a' is colored yellow
```

```python
#on pure raw data
plt.scatter(X_in['x'],X_in['y'],c=y_with_only_b)

# the c is used for color, c=y_with_only_b it
# means that color the data that is inside of variable
# and we know that this variable contains only 'b' point
# you can see results in graph that 'b' is colored yell
```

### _In the above scatter plot you can observe outliers_

## _Now we will use two different data preprocessing techniques to check which one suits our dataset better_

1. StandardScaler()
2. QuantileTransform()

```python
scaled_X = StandardScaler().fit_transform(X_in)
```

### _Drawing scatter plot on transformed data_

```python
plt.scatter(scaled_X[:,0], scaled_X[:,1], c=y_with_only_a)
```

### _we can see from above graph that outliers still have great effect so this method._
### **_Hence, it is useless_**

## _Lets try Quantile transformer_

```python
scaled_X1 =  QuantileTransformer().fit_transform(X_in)
```

### _plot after Quantile transformer_

```python
plt.scatter(scaled_X1[:,0],scaled_X1[:,1],c=y_with_only_a)
```

## _Now as you can see Quantile transformers is working perfectly fine because it is made to handle outliers_
