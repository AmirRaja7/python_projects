## _for loop_

```python
for x in range(5):
    print(x)
```

```python
week = ["mon", "tue", "wed", "thur", "fri", "sat", "sun"]
for w in week:
    print(w)
```

```python
color= ['red', 'amber', 'green', 'black', 'white', 'blue']
fruit= ['banana', 'orange', 'apple', 'guava']
for x in color:
    for y in fruit:
        print(x,y)
```

```python
for x in color:
    for y in fruit:
        print(y,x)
```
