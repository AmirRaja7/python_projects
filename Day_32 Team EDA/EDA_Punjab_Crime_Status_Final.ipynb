#### 7.1 Univariate Analysis:
##### 7.1.1 **Checking the normal distribution**


```python
df.head()
```

- For Population

```python
sns.histplot(x='Population',data=df)
print('Skewness and Kurtosis Before Normalization')
df['Population'].skew(),df['Population'].kurtosis()
```

**Interpetation**

The column is left skewed or positively skewed and have less skewness.

```python
cbrt_transformation = np.cbrt(df['Population'])
sns.histplot(cbrt_transformation)
```

```python
Skewness,Kurtosis=cbrt_transformation.skew(),cbrt_transformation.kurtosis()
print(' Skewness after Normalization =', Skewness)
print(' Kurtosis after Normalization =', Kurtosis)
```

```python
#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms
axs[0].hist(df['Population'], edgecolor='black')
axs[1].hist(cbrt_transformation, edgecolor='black')
#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Transformed Data')
```

- For Crime Counts

```python
sns.histplot(x='CrimeCount',data=df)
print('Skewness and Kurtosis Before Normalization')
df['CrimeCount'].skew(),df['CrimeCount'].kurtosis()
```

```python
log_transformation=np.log(df['CrimeCount'])
sns.histplot(log_transformation)
```

```python
Skewness_Crime,Kurtosis_Crime=log_transformation.skew(),log_transformation.kurtosis()
print(' Skewness after Normalization =', Skewness_Crime,'/0')
print(' Kurtosis after Normalization =', Kurtosis_Crime,'/0')
```

### **b) Numeric vs Categoric :**

When one variable is of numeric type and another is a categorical variable, then you perform numeric-categorical analysis.
You can use the groupby function to arrange the data into similar groups. Rows that have the same value in a particular column will be arranged in a group together. This way, you can see the numerical occurrences of a certain category across a column. You can groupby values and find their mean.

- Population vs Year Analysis:

```python
# Population vs Year
plt.figure(figsize=(10, 7))
ax = sns.barplot(data=df, x='Year', y='Population')
plt.xticks(size=10)
plt.xlabel('Year', size=14)
plt.yticks(size=12)
plt.ylabel('Population', size=12)
plt.title("Population rise in every year", size=16)
for i in ax.containers:
    ax.bar_label(i,)
```

- Population vs Division Analysis:

```python
plt.figure(figsize=(15, 7))
ax = sns.barplot(data=df, x='Division', y='Population')
plt.xticks(size=10)
plt.xlabel('Division', size=14)
plt.yticks(size=12)
plt.ylabel('Populationper_year', size=12)
plt.title("Population in every division", size=16)
for i in ax.containers:
    ax.bar_label(i,)
```

- Population vs District Analysis:

```python
plt.figure(figsize=(15, 12))
ax = sns.barplot(data=df, y='District', x='Population')
plt.xticks(size=10)
plt.xlabel('Population per year ', size=14)
plt.yticks(size=12)
plt.ylabel('District', size=12)
plt.title("Population rise in every district", size=16)
```

```python
plt.figure(figsize=(15, 12))
ax = sns.boxplot(y=df['District'],x=df['Population'])
plt.xticks(size=10)
plt.xlabel('Population ', size=14)
plt.yticks(size=12)
plt.ylabel('District', size=12)
plt.title("Population rise in every district", size=16)
```

- Population vs CrimeType Analysis:

```python
plt.figure(figsize=(15, 12))
ax = sns.barplot(data=df, x='CrimeType', y='Population',hue='Division')
plt.xticks(size=10)
plt.xlabel('CrimeType ', size=14)
plt.yticks(size=12)
plt.ylabel('Population', size=12)
plt.title("Population/CrimeType", size=16)
```

```python
plt.figure(figsize=(15, 10))
ax = sns.boxplot(y=df['CrimeType'],x=df['Population'])
plt.xticks(size=10)
plt.xlabel('Population ', size=14)
plt.yticks(size=12)
plt.ylabel('CrimeType', size=12)
plt.title("Population vs CrimeType", size=16)
```
